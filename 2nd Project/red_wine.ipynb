```python
from scipy import stats
```

```python
F, p = stats.f_oneway(rw_df[rw_df['quality_label'] == 'low']['alcohol'],
rw_df[rw_df['quality_label'] == 'medium']['alcohol'],
rw_df[rw_df['quality_label'] == 'high']['alcohol'])
print('ANOVA test for mean alcohol levels across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```
